```python
#1.
pop_10 =data.loc[(data['Population'] >= 9000000) & (data['Population'] <= 11000000)]
print(list(pop_10.index))
```

```python
#2.
mean_pop_10 = pop_10['GDP'].mean()
std_pop_10 = pop_10['GDP'].std()

print(f' The average is {round(mean_pop_10,2)} and the standard deviation is {round(std_pop_10,2)}\n')
```

```python
#3.
gdp_pc = pop_10['GDP']/pop_10['Population']
gdp_pc_mean = gdp_pc.mean()
gdp_pc_std = gdp_pc.std()
print(round(gdp_pc_mean,2))
print(round(gdp_pc_std,2))
```

```python
#4.
data.loc[data['Population'] < 1000000, 'Country size'] = 'Tiny'
data.loc[(data['Population'] >= 1000000) & (data['Population'] < 5000000), 'Country size'] = 'Very Small'
data.loc[(data['Population'] >= 5000000) & (data['Population'] < 15000000), 'Country size'] = 'Small'
data.loc[(data['Population'] >= 15000000) & (data['Population'] < 30000000), 'Country size'] = 'Medium'
data.loc[(data['Population'] >= 30000000) & (data['Population'] < 100000000), 'Country size'] = 'Large'
data.loc[data['Population'] >= 100000000, 'Country size'] = 'Huge'

data['gdp_data'] = data['GDP']/data['Population']

data.groupby('Country size').agg({'gdp_data': ['mean','std']})
```

```python
#5.
def mean_bin(df, column):
    
    mean_bin = []
    total_mean = df[column].mean()
    print(total_mean)
    
    for i in range(len(df)):
        if df[column][i] < total_mean:
            mean_bin.append(0)
        elif df[column][i] > total_mean:
            mean_bin.append(1)
        else:
            mean_bin.append(None)    
    return pd.Series(mean_bin)

mean_bin(data, 'Population')
```

```python
#6.

hc_binary = mean_bin(data,'Human Capital Index')

data.insert(6,'hc_binary',list(hc_binary))

data.groupby(['Country size','hc_binary'])['gdp_data'].mean()
```

```python
#7.
gini_binary = mean_bin(data,'GINI')

data.insert(7,'gini_binary',list(gini_binary))

data.groupby(['hc_binary','gini_binary'])['gdp_data'].mean()
```

```python
#8.

data_highest = data.loc[(data['gini_binary'] ==0.0) & (data['hc_binary'] == 0.0)]

max_gdp= data_highest['gdp_data'].max()

data_highest.loc[data_highest['gdp_data'] == max_gdp]

index_code = list(data_highest.loc[data_highest['gdp_data'] == max_gdp].index)

for i in range(len(wdi)):
    if wdi['Country Code'][i] == index_code[0]:
        valor = wdi['Country Name'][i]
        print(valor)
        break
```

```python
#9.

data_zero = data.loc[(data['hc_binary'] == 0) & (data['gini_binary'] == 0 )]
#print(data_zero)

data_size = data_zero.groupby('Country size')['gdp_data'].max()
data_code =  data_zero.groupby('Country size')['gdp_data'].idxmax()

pais_max_size=data_size.idxmax()
#print(pais_max_size)
pais_code = data_code[pais_max_size]
#print(pais_code)

for i in range(len(wdi)):
    if wdi['Country Code'][i] == pais_code:
        valor = wdi['Country Name'][i]
        print(valor)
        break
```

```python
#10.

#create a new df using the columns Country name, 1980 and 2010 of wdi and getting the code from GDP

data_increase = wdi[['Country Name', '1980', '2010']][wdi['Indicator Name'].str.contains('GDP (current US$)', regex = False)]
data_increase

#change the column Country Name to index 
data_increase = data_increase.set_index('Country Name')

#formula of increase
data_increase['Increase GDP'] = ((data_increase['2010'] - data_increase['1980']) / data_increase['1980'])*100

#using the fuction max to find the max value
max_increase = data_increase['Increase GDP'].max()


#print the name of the country with max increase between 1980 and 2010.
print(list(data_increase[data_increase['Increase GDP'] == max_increase].index.values))

```
